```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CSV Files/covid_vaccine_statewise.csv')
```

```python
df.columns = df.columns.str.strip()  # Remove whitespace in column names
```

```python
#Drop rows with missing AEFI
df = df[df['AEFI'].notnull()]
```

```python
#AEFI Per Gender
df['AEFI_Male'] = (df['Male(Individuals Vaccinated)'] / df['Total Individuals Vaccinated']) * df['AEFI']
df['AEFI_Female'] = (df['Female(Individuals Vaccinated)'] / df['Total Individuals Vaccinated']) * df['AEFI']
df['AEFI_Transgender'] = (df['Transgender(Individuals Vaccinated)'] / df['Total Individuals Vaccinated']) * df['AEFI']
```

```python
#Declaring AEFI as high or low, so to do that we use the median, and anything less is low and greater is high
threshold = df['AEFI'].median()
df['AEFI_High'] = (df['AEFI'] > threshold).astype(int)
```

```python
features = [
    'Male(Individuals Vaccinated)',
    'Female(Individuals Vaccinated)',
    'Transgender(Individuals Vaccinated)',
    'Total Individuals Vaccinated'
]
X = df[features]
y = df['AEFI_High']
```

```python
sns.pairplot(df[features + ['AEFI_High']], hue='AEFI_High')
plt.show()
```

```python
#Training AEFI = Adverse Events Following Immunization
```

```python
X = df[features]
y = df['AEFI_High']

# Drop rows where any feature is NaN (As GuassianNB works only with numbers)
mask = X.notnull().all(axis=1)
X = X[mask]
y = y[mask]
```

```python
print(df['AEFI_High'].value_counts()) 
#Ensuring values are comparable
```

```python
from sklearn.model_selection import train_test_split

X = df[[
    'Male(Individuals Vaccinated)',
    'Female(Individuals Vaccinated)',
    'Transgender(Individuals Vaccinated)',
    'Total Individuals Vaccinated'
]].fillna(0)

y = df['AEFI_High']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Low AEFI', 'High AEFI'], 
            yticklabels=['Low AEFI', 'High AEFI'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - AEFI Classification')
plt.show()
```
